# 01 Data Exploration

This notebook explores the four Eniac datasets, checks structure, identifies potential data quality issues, and prepares the project roadmap.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 120)

# File paths
orders_path = '/mnt/data/orders_cl.csv'
products_path = '/mnt/data/products_cl(1).csv'
orderlines_path = '/mnt/data/orderlines_qu.csv'
brands_path = '/mnt/data/brands(1).csv'

# Load datasets
orders_cl = pd.read_csv(orders_path)
products_cl = pd.read_csv(products_path)
orderlines_qu = pd.read_csv(orderlines_path)
brands_df = pd.read_csv(brands_path)

print('orders_cl:', orders_cl.shape)
print('products_cl:', products_cl.shape)
print('orderlines_qu:', orderlines_qu.shape)
print('brands_df:', brands_df.shape)
```

## 1. Dataset overview

```
for name, df in [('orders_cl', orders_cl), ('products_cl', products_cl), ('orderlines_qu', orderlines_qu), ('brands_df', brands_df)]:
    print(f'\n{name}')
    display(df.head())
    print(df.info())
```

## 2. Missing values

```
missing_summary = {}
for name, df in [('orders_cl', orders_cl), ('products_cl', products_cl), ('orderlines_qu', orderlines_qu), ('brands_df', brands_df)]:
    summary = pd.DataFrame({
        'missing_count': df.isna().sum(),
        'missing_percent': (df.isna().mean() * 100).round(2)
    })
    missing_summary[name] = summary
    print(f'\n{name}')
    display(summary)
```

## 3. Duplicate checks

```
print('Duplicate rows in orders_cl:', orders_cl.duplicated().sum())
print('Duplicate rows in products_cl:', products_cl.duplicated().sum())
print('Duplicate rows in orderlines_qu:', orderlines_qu.duplicated().sum())
print('Duplicate rows in brands_df:', brands_df.duplicated().sum())

print('\nDuplicate SKUs in products_cl:', products_cl['sku'].duplicated().sum())
print('Duplicate order_id in orders_cl:', orders_cl['order_id'].duplicated().sum())
print('Duplicate brand codes in brands_df:', brands_df['short'].duplicated().sum())
```

## 4. Key integrity checks

```
# Check whether all sold SKUs exist in the product catalog
orphan_skus = orderlines_qu.loc[~orderlines_qu['sku'].isin(products_cl['sku']), 'sku'].nunique()

# Check whether all orderlines belong to a known order
orphan_orders = orderlines_qu.loc[~orderlines_qu['id_order'].isin(orders_cl['order_id']), 'id_order'].nunique()

print('Sold SKUs not found in products_cl:', orphan_skus)
print('Order IDs in orderlines_qu not found in orders_cl:', orphan_orders)
```

## 5. Exploratory distributions

```
display(orders_cl['state'].value_counts(dropna=False))
display(products_cl['price'].describe())
display(orderlines_qu['unit_price'].describe())
display(orderlines_qu['product_quantity'].describe())
```
